```python
import gspread
import pandas as pd
```

```python
service_account_creds_path = "Google_Api_Credentials/service_account.json"
```

```python
service_account = gspread.service_account(filename=service_account_creds_path)
```

```python
spreadsheet = service_account.open_by_url("https://docs.google.com/spreadsheets/d/1ssxsUKICWBO2p07SoX6bEVqT_DVkXmqovyWZhuOFP7s/edit#gid=0")
```

```python
spreadsheet.worksheets()
```

```python
responses_sheet = spreadsheet.get_worksheet(0)
```

```python
df = pd.DataFrame(responses_sheet.get_all_values())
df = df.T.set_index(0).T
df
```

```python
df_pairs = df[df['One or pair?']=='Pair']
# df_pairs = df_pairs[["Mon 1", "Mon 2"]].apply(lambda x:x.str.strip())
# df_pairs.dropna(inplace=True)
df_pairs_but_only_1 = pd.DataFrame([row[1] for row in df_pairs.iterrows() if any([row[1]['Mon 1'].isspace(),row[1]['Mon 2'].isspace()])]).reset_index(drop=True)
df_pairs = pd.DataFrame([row[1] for row in df_pairs.iterrows() if not any([row[1]['Mon 1'].isspace(),row[1]['Mon 2'].isspace()])])
df_pairs.reset_index(drop=True,inplace=True)
display(df_pairs)
display(df_pairs_but_only_1)
```

```python
df_single_ideas = df[df['One or pair?']=='One']
df_single_ideas = df_single_ideas[['Pokemon']]
df_single_ideas = df_single_ideas[df_single_ideas['Pokemon'].map(lambda x: x.isspace() is False)]
df_single_ideas.reset_index(drop=True,inplace=True)
alex_ideas = df_single_ideas.iloc[df_single_ideas.index[1::2]]
jonny_ideas = df_single_ideas.iloc[df_single_ideas.index[::2]]

jonny_sheet = spreadsheet.worksheet("Jonny")
alex_sheet = spreadsheet.worksheet("Alex")

jonny_sheet.append_rows(jonny_ideas.values.tolist())
alex_sheet.append_rows(alex_ideas.values.tolist())
```
